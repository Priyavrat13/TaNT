```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
from PIL import Image
from tensorflow.keras.preprocessing import image
```

```python
dir_path = 'test_images/' #path to images
```

```python
TaNT = tf.keras.models.load_model('TaNT.h5') #loading TaNT classification model
```

```python
for i in os.listdir(dir_path):
    img = image.load_img(dir_path + '//' + i, target_size=(224,224,3))
    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    images = np.vstack([X])
    prediction = TaNT.predict(images)
    predicted_id = tf.math.argmax(prediction, axis=-1)
    print(i)
    if predicted_id ==2:
        plt.imshow(img)
        plt.show()
        print("Image is Textual")
    elif predicted_id ==1:
        plt.imshow(img)
        plt.show()
        print("Image is Non-Textual")
    elif predicted_id ==0:
        plt.imshow(img)
        plt.show()
        print("Image have combined content")
```
